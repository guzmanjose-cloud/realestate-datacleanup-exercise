#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO
normalized_subset = ds_subset.copy()
normalized_subset['normalized_price'] = normalized_subset.groupby('level5')['price'].transform(lambda x: (x - x.mean()) / x.std())

plt.figure(figsize=(10, 6))
for population, group in normalized_subset.groupby('level5'):
    plt.hist(group['normalized_price'], bins=20, alpha=0.5, label=population)

plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized Prices for Each Population')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in the question 14) (★★★★)

Execute the following cell and then start coding in the next one. You must implement a simple code that transform the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps, MarkerCluster, Marker
from ipyleaflet import AwesomeIcon

import random

map = Map(center=(60.0, -2.2), zoom=2, min_zoom=1, max_zoom=10, basemap={
    "url": "https://stamen-tiles-{s}.a.ssl.fastly.net/terrain/{z}/{x}/{y}{r}.png",
    "attribution": "&copy; <a href='https://maps.stamen.com/terrain'>Stamen</a> contributors"
})

```

```python

level_colors = {
    'Fuenlabrada': 'blue',
    'Alcorcón': 'red',
    'Leganés': 'green',
    'Getafe': 'black'
}

markers = []

for index, row in ds_subset.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    population = row['level5']
    
    color = level_colors.get(population, 'black')
    
    icon = AwesomeIcon(name='circle', marker_color=color, icon_color='white', spin=False)
    
    marker = Marker(location=(latitude, longitude), draggable=False, title=population, icon=icon)
    
    markers.append(marker)

marker_cluster = MarkerCluster(markers=markers)

map.add_layer(marker_cluster)

map
```
